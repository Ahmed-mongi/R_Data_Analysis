# **Hypothesis Testing**

```r
# Load required package
if(!require(stats)) install.packages('stats')

# Assuming Outcome 1 means diabetic and 0 means non-diabetic
diabetic_glucose <- data$Glucose[data$Outcome == 1]
non_diabetic_glucose <- data$Glucose[data$Outcome == 0]

# Perform the independent samples t-test
t_test_result <- t.test(diabetic_glucose, non_diabetic_glucose, alternative = "two.sided")

# Print the test results
print(t_test_result)
```

***Given the p-value is much smaller than 0.05, we reject the null hypothesis. There is strong evidence to support the claim that there is a significant difference in glucose levels between diabetic and non-diabetic patients. The results suggest that diabetic patients have higher average glucose levels compared to non-diabetic patients, which aligns with medical expectations and the nature of diabetes as a condition characterized by elevated blood glucose levels.***

```r
# Claim: "There is a significant difference in BMI between diabetic and non-diabetic patients."
diabetic_bmi <- data$BMI[data$Outcome == 1]
non_diabetic_bmi <- data$BMI[data$Outcome == 0]

# Perform the independent samples t-test
bmi_t_test_result <- t.test(diabetic_bmi, non_diabetic_bmi, alternative = "two.sided")

# Print the test results
print(bmi_t_test_result)
```

***Given the p-value is much smaller than 0.05, we reject the null hypothesis.
There is strong evidence to support the claim that there is a significant difference in BMI between diabetic and non-diabetic patients. The results suggest that diabetic patients have a higher average BMI compared to non-diabetic patients. This finding aligns with the understanding that higher BMI is often associated with an increased risk of developing diabetes.***

# **Simulation Task**

```r
glucose_data <-data$Glucose

# Calculate the true population mean
true_mean <- mean(glucose_data, na.rm = TRUE)
print(paste("True Population Mean of Glucose:", round(true_mean, 2)))
```

```r
# Set seed for reproducibility
set.seed(123)

# Function to compute 95% CI for a sample
compute_ci <- function(sample) {
  sample_mean <- mean(sample)
  sample_sd <- sd(sample)
  n <- length(sample)
  error_margin <- qt(0.975, df = n - 1) * (sample_sd / sqrt(n))
  lower_bound <- sample_mean - error_margin
  upper_bound <- sample_mean + error_margin
  return(c(mean = sample_mean, lower = lower_bound, upper = upper_bound))
}

# 4.1: 25 Samples of Size 15
sample_size_15 <- 15
num_samples_15 <- 25

# Initialize a dataframe to store results
results_15 <- data.frame(
  Sample = integer(),
  Sample_Mean = double(),
  CI_Lower = double(),
  CI_Upper = double(),
  Contains_True_Mean = logical()
)

for (i in 1:num_samples_15) {
  sample <- sample(glucose_data, size = sample_size_15, replace = FALSE)
  ci <- compute_ci(sample)
  contains <- (ci['lower'] <= true_mean) & (true_mean <= ci['upper'])

  results_15 <- results_15 %>%
    add_row(
      Sample = i,
      Sample_Mean = ci['mean'],
      CI_Lower = ci['lower'],
      CI_Upper = ci['upper'],
      Contains_True_Mean = contains
    )
}

# Display the results
print(results_15)

# Calculate the coverage proportion
coverage_15 <- mean(results_15$Contains_True_Mean)
print(paste("Coverage Proportion for Sample Size 15:", round(coverage_15 * 100, 2), "%"))
```

```r
# 4.2: 25 Samples of Size 100
sample_size_100 <- 100
num_samples_100 <- 25

# Initialize a dataframe to store results
results_100 <- data.frame(
  Sample = integer(),
  Sample_Mean = double(),
  CI_Lower = double(),
  CI_Upper = double(),
  Contains_True_Mean = logical()
)

for (i in 1:num_samples_100) {
  sample <- sample(glucose_data, size = sample_size_100, replace = FALSE)
  ci <- compute_ci(sample)
  contains <- (ci['lower'] <= true_mean) & (true_mean <= ci['upper'])

  results_100 <- results_100 %>%
    add_row(
      Sample = i,
      Sample_Mean = ci['mean'],
      CI_Lower = ci['lower'],
      CI_Upper = ci['upper'],
      Contains_True_Mean = contains
    )
}

# Display the results
print(results_100)

# Calculate the coverage proportion
coverage_100 <- mean(results_100$Contains_True_Mean)
print(paste("Coverage Proportion for Sample Size 100:", round(coverage_100 * 100, 2), "%"))

# Compare CI widths
ci_width_15 <- results_15$CI_Upper - results_15$CI_Lower
ci_width_100 <- results_100$CI_Upper - results_100$CI_Lower

average_width_15 <- mean(ci_width_15)
average_width_100 <- mean(ci_width_100)

print(paste("Average CI Width for Sample Size 15:", round(average_width_15, 2)))
print(paste("Average CI Width for Sample Size 100:", round(average_width_100, 2)))
```

```r
# 4.3: 20 Samples of Size 10
sample_size_10 <- 10
num_samples_10 <- 20

# Initialize a dataframe to store results
results_10 <- data.frame(
  Sample = integer(),
  Sample_Mean = double(),
  CI_Lower = double(),
  CI_Upper = double(),
  Contains_True_Mean = logical()
)

for (i in 1:num_samples_10) {
  sample <- sample(glucose_data, size = sample_size_10, replace = FALSE)
  ci <- compute_ci(sample)
  contains <- (ci['lower'] <= true_mean) & (true_mean <= ci['upper'])

  results_10 <- results_10 %>%
    add_row(
      Sample = i,
      Sample_Mean = ci['mean'],
      CI_Lower = ci['lower'],
      CI_Upper = ci['upper'],
      Contains_True_Mean = contains
    )
}

# Display the results
print(results_10)

# Calculate the coverage proportion
coverage_10 <- mean(results_10$Contains_True_Mean)
print(paste("Coverage Proportion for Sample Size 10:", round(coverage_10 * 100, 2), "%"))

# Compare CI widths with previous sample sizes
ci_width_10 <- results_10$CI_Upper - results_10$CI_Lower

average_width_10 <- mean(ci_width_10)

print(paste("Average CI Width for Sample Size 10:", round(average_width_10, 2)))
```

```r
# Combine all results into one dataframe with Sample Size labels
results_15$Sample_Size <- 15
results_100$Sample_Size <- 100
results_10$Sample_Size <- 10

combined_results <- bind_rows(results_15, results_100, results_10)

# Plot CI Widths by Sample Size
ggplot(combined_results, aes(x = factor(Sample_Size), y = CI_Upper - CI_Lower, fill = factor(Sample_Size))) +
  geom_boxplot() +
  labs(title = "Comparison of Confidence Interval Widths by Sample Size",
       x = "Sample Size",
       y = "CI Width",
       fill = "Sample Size") +
  theme_minimal()
```

```r
# Create a summary dataframe for coverage proportions
coverage_summary <- data.frame(
  Sample_Size = c(10, 15, 100),
  Coverage = c(coverage_10, coverage_15, coverage_100)
)

# Plot Coverage Proportion
ggplot(coverage_summary, aes(x = factor(Sample_Size), y = Coverage * 100, fill = factor(Sample_Size))) +
  geom_bar(stat = "identity") +
  geom_hline(yintercept = 95, linetype = "dashed", color = "red") +
  labs(title = "Coverage Proportion by Sample Size",
       x = "Sample Size",
       y = "Coverage Proportion (%)",
       fill = "Sample Size") +
  theme_minimal()
```
